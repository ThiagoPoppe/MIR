# Quest√£o 04)

```python
import numpy as np
from scipy.signal import chirp
import matplotlib.pyplot as plt

def generate_chirp_signal(f0, ft, T):
    fs = 2.5 * ft
    n_samples = int(fs * T)
    
    t = np.linspace(0, T, n_samples)
    x = chirp(t, f0=f0, t1=T, f1=ft, method='linear')
    
    return x, t

signal, t = generate_chirp_signal(f0=15, ft=150, T=1000)

plt.xlabel('Time [sec]')
plt.plot(t, signal)
plt.show()
```

```python
from scipy.fft import fft
fft_values = np.abs(fft(signal))

plt.title('FFT of chirp signal')
plt.xlabel('Frequency (index k)')
plt.plot(fft_values)
plt.show()
```

# Quest√£o 05)

```python
from scipy.signal import stft
f, t, Zxx = stft(signal, window='hann', nperseg=10)

plt.figure(figsize=(10, 5))
plt.pcolormesh(t, f, np.log(np.abs(Zxx)), shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
